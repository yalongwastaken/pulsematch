```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
def plot_accuracies(
      results: list[dict[str, any]],
      categories: list[str]=['Original Accuracy', 'Weighted Accuracy', 'Non-zero Tap Accuracy']
      ) -> None:
    plt.figure(figsize=(16, 8))

    thresholds = np.linspace(0.10, 0, num=21)
    colors = plt.get_cmap(name='tab10')

    markers = {
        "Original Accuracy": 'o',
        "Weighted Accuracy": 's',
        "Non-zero Tap Accuracy": '^'
    }

    # Legend for shapes (accuracy types)
    shape_handles = []

    for idx, result in enumerate(results):
        scenario = result["Scenario"]
        color = colors(idx)

        if "Original Accuracy" in categories:
          original_accuracy = result["Original_Accuracy"]
          plt.plot(thresholds, original_accuracy, marker=markers["Original Accuracy"],
                   color=color, linestyle='-', label=scenario if idx == 0 else None)
          if idx == 0:
            shape_handles.append(Line2D([0], [0], marker='o', color='gray', label='Original Accuracy', linestyle=''))

        if "Weighted Accuracy" in categories:
          weighted_accuracy = result["Weighted_Accuracy"]
          plt.plot(thresholds, weighted_accuracy, marker=markers["Weighted Accuracy"],
                 color=color, linestyle='--')
          if idx == 0:
            shape_handles.append(Line2D([0], [0], marker='s', color='gray', label='Weighted Accuracy', linestyle=''))

        if "Non-zero Tap Accuracy" in categories:
          nonzero_tap_accuracy = result["Nonzero_Tap_Accuracy"]
          plt.plot(thresholds, nonzero_tap_accuracy, marker=markers["Non-zero Tap Accuracy"],
                 color=color, linestyle=':')
          if idx == 0:
            shape_handles.append(Line2D([0], [0], marker='^', color='gray', label='Non-zero Tap Accuracy', linestyle=''))

    # Legend for colors (scenarios)
    color_handles = [
        Line2D([0], [0], color=colors(idx), lw=3, label=result["Scenario"])
        for idx, result in enumerate(results)
    ]

    plt.xlabel('Threshold', fontsize=12)
    plt.ylabel('Threshold Accuracy Percentage (%)', fontsize=12)
    plt.title('Threshold Accuracy Percentage vs. Threshold', fontsize=14)
    plt.gca().invert_xaxis()
    plt.xticks(np.round(np.arange(0.10, -0.001, -0.01), 3), fontsize=10, rotation=-45)
    plt.grid(True, linestyle='--', linewidth=0.7)

    # Legends
    legend1 = plt.legend(handles=shape_handles, title="Accuracy Type", loc='lower left', fontsize=9)
    legend2 = plt.legend(handles=color_handles, title="Scenario", loc='lower right', fontsize=9)
    plt.gca().add_artist(legend1)

    plt.tight_layout()
    plt.show()
```

```python
results = [scenario_1_results, scenario_2_results]

plot_accuracies(results)
```
